## <font color='green'> <b>Create a 2d array with 1 on the border and 0 inside</b><font color='black'>

```python
rows, cols = 5, 5 

array = np.zeros((rows, cols), dtype=int)

array[0, :] = 1       # Top row
array[-1, :] = 1      # Bottom row
array[:, 0] = 1       # Left column
array[:, -1] = 1      # Right column

print(array)
```
